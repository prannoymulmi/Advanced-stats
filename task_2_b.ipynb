```python
import numpy as np
import matplotlib.pyplot as plt


# Definitions
# both 500 points
x = np.arange(0, 360, 0.072)
y = np.arange(0, 1, 0.0002)
z = np.repeat(1 / 360, 5000)

# Plot the probability distribution
plt.subplots(1, 1, figsize=(20, 10))
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'gray')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
ax.scatter3D(x, y, z, c=z, cmap='coolwarm_r');


plt.savefig('images/task_2_b.svg')
```
